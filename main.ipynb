```python
from tkinter import *
import requests

FONT = ('Courier', 24, 'bold')

def yeezy_quotes():
    response = requests.get('https://api.kanye.rest/')
    response.raise_for_status()
    data = response.json()
    quote = data['quote']
    canvas.itemconfig(quote_text, text = quote)

window = Tk()
window.title('Daily Kanye-Isms...')
window.config(padx = 50, pady = 50)

canvas = Canvas(width = 300, height = 425)
back_img = PhotoImage(file = 'background.png')
yeezy_img = PhotoImage(file = 'kanye.png')
canvas.create_image(150, 212, image = back_img)
quote_text = canvas.create_text(150, 212, text = '', width = 250, font = FONT, fill = 'white')
canvas.grid(row = 0, column = 0)
yeezy_button = Button(image = yeezy_img, highlightthickness = 0, command = yeezy_quotes)
yeezy_button.grid(row = 1, column = 0)

window.mainloop()
```
